```python
#10) Create a new column in movies_df that catagorizes each movie as either
#short, average, or long.  Short movies should be 30 mins or less, long movies
#should be over 2 hours.
```

```python
movie_df['duration'] = movie_df['duration'].str.replace('min', '', regex=False).str.strip().astype(int)
movie_df.head()
```

```python
movie_df['time_investment'] = ''
```

```python
for ind, row in movie_df.iterrows():
    if row['duration'] <= 30:
        movie_df.loc[ind, 'time_investment'] = 'short'
    elif row['duration'] > 120:
        movie_df.loc[ind, 'time_investment'] ='long'
    else:
        movie_df.loc[ind, 'time_investment'] ='middle'
```

```python
movie_df
```

```python
cat_counts = movie_df.time_investment.value_counts()
cat_counts
```

```python
plt.pie(cat_counts['count'], labels = cat_counts['time_investment'])
```
