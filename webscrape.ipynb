```python
import requests
```

```python
URL = 'https://realpython.github.io/fake-jobs/'
```

```python
response = requests.get(URL)
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.find('title')
```

```python
import pandas as pd
```

```python
soup.find('title').text
```

```python
#can you find a tag type and/or a class that could be helpful for extracting this information? Extract the text from this title.
#</h1>
```

```python
jobtitlestxt = soup.find_all("h2")
for jobtitlestxt in jobtitlestxt:
    print(jobtitlestxt.text)
```

```python
jobtitles = soup.findAll('h2')
print(type(jobtitles))
jobtitles
```

```python
jobtitleslist = [x.text.strip() for x in jobtitles]  #x.text extracted the text inside the <h2> tags, and .strip() got rid 
#of extra space or characters    .strip() cleans it up 
jobtitleslist                                
```

```python
#extract the companies, locations, and posting dates ---"Payne, Roberts and Davis", a location of "Stewartbury, AA", 
#and a posting date of "2021-04-08"
```

```python
company = soup.findAll('h3')
print(type(company))
company
```

```python
companylist = [x.text.strip() for x in company] 
companylist
```

```python
location = soup.findAll(class_= 'location')
print(type(location))
location
```

```python
locationlist = [x.text.strip() for x in location] 
locationlist
```

```python
time = soup.findAll('time')
print(type(time))
time
```

```python
timelist = [x.text.strip() for x in time] 
timelist
```

```python
newforcejobs = pd.DataFrame({'Job':jobtitleslist, 'Location': locationlist, 'Date Posted': timelist})
newforcejobs
```
